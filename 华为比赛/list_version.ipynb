```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
PATH = './output/version'
```

```python
def list_columns(num):
    for i in np.arange(1, num):
        print('Version{}'.format(i))
        print('-' * 100)
        train = pd.read_csv(PATH + str(i) + '/train_data.csv', nrows = 10)
        print(train.columns.values)
        print('the length is {}'.format(len(train.columns)))
        print('\n\n\n')
```

```python
list_columns(5)
```
