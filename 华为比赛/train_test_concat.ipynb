```python
import pandas as pd
import numpy as np
```

```python
age_train = pd.read_csv("age_train.csv", names=['uid','age_group'])
age_test = pd.read_csv("age_test.csv", names=['uid'])
user_basic_info = pd.read_csv("user_basic_info.csv", names=['uid','gender','city','prodName','ramCapacity','ramLeftRation','romCapacity','romLeftRation','color','fontSize','ct','carrier','os'])
user_behavior_info = pd.read_csv("user_behavior_info.csv", names=['uid','bootTimes','AFuncTimes','BFuncTimes','CFuncTimes','DFuncTimes','EFuncTimes','FFuncTimes','FFuncSum'])
```

```python
uid_used_set = set(age_train.uid.values.tolist())
```

```python
uid_used_set = list(uid_used_set)
```

```python
user_info = pd.DataFrame()
user_info['uid'] = uid_used_set
```

```python
user_info = user_info.merge(user_basic_info, how='left', on='uid', suffixes=('_base0', '_ubaf'))
user_info = user_info.merge(user_behavior_info, how='left', on='uid', suffixes=('_base1', '_ubef'))
```

**师姐写错了carrier**

```python
cat_columns = ['city','prodName','color','carrier','ct','carrier']
```

```python
user_info[cat_columns] = user_info[cat_columns].astype(str)
```

```python
for cat in cat_columns:
    temp = np.unique(user_info[cat].values)
    temp_dict = dict(zip(temp, np.arange(1, len(temp)+1)))
    user_info[cat + '_num'] = user_info[cat].apply(lambda x: temp_dict[x]) 
```
